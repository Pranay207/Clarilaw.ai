```python
crpc['text_length'] = crpc['Description'].astype(str).apply(len)

print("CRPC Average text length:", crpc['text_length'].mean())
```

```python
final_ic['text_length'] = final_ic['article_desc'].astype(str).apply(len)

print("Final_IC Average text length:", final_ic['text_length'].mean())
```
